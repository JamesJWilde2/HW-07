### Part 2

Scrape the content of https://www.lemonde.fr/ and save it as a CSV.

We want: titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

response = requests.get('https://www.lemonde.fr/')
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
articles = doc.find_all(class_='article')
len(articles)
```

```python
rows = []

for article in articles:
    row = {}

    row['title'] = article.find(class_='article__title').text

    try:
        row['subhead'] = article.find(class_='article__desc').text
    except:
        pass

    try:
        row['url'] = article.find('a').get('href', None)
    except:
        row['url'] = article.get('href', None)

    try:
        row['is_premium'] = article.find(class_='icon__premium').text
    except:
        pass
        
    try:
        row['image_url'] = article.select_one('img[src]').get('src', None)
    except:
        pass

    try:
        row['byline'] = article.find(class_='article__author-name').text
    except:
        pass

    try:
        row['article type'] = article.find(class_='article__type').text
    except:
        pass

    rows.append(row)
    
rows
```

```python
df = pd.DataFrame(rows)
df.to_csv('lemonde_articles.csv', index=False)
```
